```python
def plotuni(var):
    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2)=plt.subplots(1,2,figsize=(20,6))
    
    sns.countplot(x=var,data=df0,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {var} for Non Defaulter',fontsize=15)
    ax1.set_xticklabels(ax1.get_xticklabels(),rotation=40,ha='right')
    #Including percentage 
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(df0))*100),(p.get_x()+0.1,p.get_height()+50))
        
    sns.countplot(x=var ,data=df1,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {var} for  Defaulter',fontsize=15)
    ax2.set_xticklabels(ax2.get_xticklabels(),rotation=40,ha='right')
    
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(df1))*100),(p.get_x()+0.1,p.get_height()+50))
        
    plt.show()
        
    
```

```python
#For continous variable

def plotunidist(var):
    plt.style.use('ggplot')
    sns.despine
    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(15,5))
    sns.distplot(a=df0[var],ax=ax1)
    ax1.set_title(f'Distribution of {var} for Non Defaulter',fontsize=15)
    
    sns.distplot(a=df1[var],ax=ax2)
    ax2.set_title(f'Distribution of {var} for  Defaulter',fontsize=15)
    plt.show()
```

```python
#Bivariate analysis of continuous variable
def plotbivar(var1,var2):
    plt.style.use('ggplot')
    sns.despine
    fig, (ax1,ax2)=plt.subplots(1,2,figsize=(15,5))
    sns.scatterplot(x=var1,y=var2,data=df0,ax=ax1)
    ax1.set_xlabel(var1)
    ax1.set_ylabel(var2)
    ax1.set_title(f' {var1} vs. {var2} for Non Defaulter',fontsize=15)
    
    sns.scatterplot(x=var1,y=var2,data=df1,ax=ax2)
    ax2.set_xlabel(var1)
    ax2.set_ylabel(var2)
    ax2.set_title(f' {var1} vs. {var2} for  Defaulter',fontsize=15)
    
    
    plt.show()
    
```

```python
#Boxplot for more bivariate analysis

def plot_catnum(cat,num):
    plt.style.use('ggplot')
    sns.despine
    fig ,ax=plt.subplots(1,1,figsize=(10,8))
    sns.boxenplot(x=cat,y=num,data=previousdf)
    ax.set_ylabel(f'{num}')
    ax.set_xlabel(f'{cat}')
    ax.set_title(f'{cat} vs.{num}',fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
    
    plt.show()
```

```python
#Bar chart

def plot_combined(varx,vary):
    plt.style.use('ggplot')
    sns.despine
    NewDat=Leftpreviousnew.pivot_table(values='SK_ID_CURR',
                                       index=varx,
                                       columns=vary,
                                       aggfunc='count')
    NewDat=NewDat.div(NewDat.sum(axis=1),axis='rows')*100
    sns.set()
    NewDat.plot(kind='bar',stacked=True,figsize=(15,5))
    plt.title(f'Impact of {varx} on loan approval')
    plt.xlabel(f'{varx}')
    plt.ylabel(f'{vary}%')
    
    plt.show()
                                       
                                       
                                       
                              
```

```python
plot_combined('FLAG_OWN_CAR','NAME_CONTRACT_STATUS')

#It shows that car owner ship does not impact on loan but earlier it was there hence more weightage should be given to it
```

```python
plot_combined('CODE_GENDER','NAME_CONTRACT_STATUS')

#There is no impact of gender however earlier female were less defaulter hence more weightage should be given to it
```

```python
plot_combined('TARGET','NAME_CONTRACT_STATUS')
#People who has already availed loan are less dafaulter
```
